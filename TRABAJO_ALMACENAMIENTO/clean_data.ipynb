```python
import pandas as pd

# 1. Leer el archivo Excel
# Asegúrate de instalar openpyxl: pip install openpyxl
df = pd.read_excel('SS2324_laliga_players.xlsx')

# 2. Convertir la columna 'Salario' a numérica
# 'coerce' convierte los errores (valores no numéricos) en NaN
df['Salario'] = pd.to_numeric(df['Salario'], errors='coerce')

# 3. Eliminar los jugadores que no tienen salario (donde Salario es NaN)
df_cleaned = df.dropna(subset=['Salario'])

# 4. Rellenar todos los demás valores NA con 0
df_cleaned = df_cleaned.fillna(0)

# Ver el resultado
print(df_cleaned.head())

# Suponiendo que tu variable con los datos limpios se llama 'df_cleaned'
df_cleaned.to_csv('SS2324_laliga_players_cleaned.csv', index=False)
```
